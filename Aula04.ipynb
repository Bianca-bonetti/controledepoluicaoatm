```python
#criando uma lista em python
list = [0, 'eu amo polatm',2.5]
print (list) # a list admite valores diferentes e trata os vetores de maneira independente
print ('variável: ' + list[1]) #adição de strings
print (list[0] + 1) #adição de números - lembrar que a indexação do python inicia em 0
```

```python
#criando uma matriz numpy
import numpy as np
a = np.zeros((5,5)) #criando uma matriz de zeros utilizando a função np.zeros(())
print(a)
print('Indexação numpy')
print(a[0,0])
```

```python
#matriz de numeros aleatórios
import numpy as np
b = np.random.rand(5,5)
print(b)
```

```python
vmin=1
vmax=100
c = vmin+(vmax-vmin)*b
print(c)
```

```python
#plotar matriz
import matplotlib.pyplot as plt
plt.plot(c)
```

```python
x = np.array(range(0,np.size(c[0])))
print(x)
plt.plot(c[0]) #plotando apenas a primeira linha
plt.xticks(x)
plt.title('Gráfico com valores Inteiros no Eixo X')
```

```python
plt.plot(c.transpose())
```

```python
#criando um gráfico de barras e exportando
fig, ax = plt.subplots(3,1) #ax é um objeto plt
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])
#exportando
import os
rootfolder = os.getcwd()
print(rootfolder)
figfolder = rootfolder.replace("\\","/")+'/figures/'
print(figfolder)
fig.savefig(figfolder+'Figura01.png')
```
